```python
import os
import numpy as np
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from PIL import Image
import matplotlib.pyplot as plt
import torchvision.transforms as T

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print("Using device:", device)

img_dir = "/content/drive/MyDrive/Nullclass internship/Task 1/Shapes Dataset"
img_files = sorted([f for f in os.listdir(img_dir) if f.lower().endswith('.png')])
img_paths = [os.path.join(img_dir, fname) for fname in img_files]
checkpoint_dir = "/content/drive/MyDrive/Nullclass internship/Task 6/Checkpoints"
os.makedirs(checkpoint_dir, exist_ok=True)

embeddings = np.load('/content/drive/MyDrive/Nullclass internship/Task 5/shape_label_embeddings.npy')
label_path = '/content/drive/MyDrive/Nullclass internship/Task 5/shape_labels.txt'
with open(label_path, 'r') as f:
    all_labels = [line.strip() for line in f]

augment = T.Compose([
    T.Resize((128, 128)),
    T.ToTensor(),
    T.Lambda(lambda x: x * 2 - 1)
])

def standardize_fg_bg(img):
    arr = np.array(img)
    if arr.mean() < 128:
        arr = 255 - arr
    return Image.fromarray(arr)

class ShapeOnlyDataset(Dataset):
    def __init__(self, img_paths, embeddings, size=(128,128), augment=None):
        assert len(img_paths) == embeddings.shape[0], "Mismatch between images and embedding count"
        self.img_paths = img_paths
        self.embeddings = embeddings
        self.size = size
        self.augment = augment
    def __len__(self):
        return len(self.img_paths)
    def __getitem__(self, idx):
        img = Image.open(self.img_paths[idx]).convert('L').resize(self.size)
        img = standardize_fg_bg(img)
        img = self.augment(img) if self.augment else T.ToTensor()(img) * 2 - 1
        embedding = torch.tensor(self.embeddings[idx], dtype=torch.float32)
        return img, embedding

batch_size = 32
dataset = ShapeOnlyDataset(img_paths, embeddings, size=(128,128), augment=augment)
loader = DataLoader(dataset, batch_size=batch_size, shuffle=True)

latent_dim = 100
embedding_dim = embeddings.shape[1]
ngf = 128
ndf = 32
img_shape = (1, 128, 128)

class ConditionalDCGANGenerator(nn.Module):
    def __init__(self, latent_dim, embedding_dim, ngf, img_shape):
        super().__init__()
        def IN2(num): return nn.InstanceNorm2d(num, affine=True)
        self.fc = nn.Sequential(
            nn.Linear(latent_dim + embedding_dim, ngf*8*4*4),
            nn.LayerNorm(ngf*8*4*4),
            nn.ReLU(True),
        )
        self.deconv = nn.Sequential(
            nn.ConvTranspose2d(ngf*8, ngf*4, 4, 2, 1),
            IN2(ngf*4), nn.ReLU(True),
            nn.ConvTranspose2d(ngf*4, ngf*2, 4, 2, 1),
            IN2(ngf*2), nn.ReLU(True),
            nn.ConvTranspose2d(ngf*2, ngf, 4, 2, 1),
            IN2(ngf), nn.ReLU(True),
            nn.ConvTranspose2d(ngf, ngf//2, 4, 2, 1),
            IN2(ngf//2), nn.ReLU(True),
            nn.ConvTranspose2d(ngf//2, img_shape[0], 4, 2, 1),
            nn.Tanh()
        )
        self.img_shape = img_shape
    def forward(self, noise, cond_emb):
        x = torch.cat([noise, cond_emb], dim=1)
        x = self.fc(x)
        x = x.view(x.size(0), ngf*8, 4, 4)
        img = self.deconv(x)
        return img

class ConditionalDCGANDiscriminator(nn.Module):
    def __init__(self, embedding_dim, ndf, img_shape):
        super().__init__()
        from torch.nn import InstanceNorm2d as IN
        self.embed_fc = nn.Linear(embedding_dim, img_shape[1]*img_shape[2])
        self.conv1 = nn.Sequential(
            nn.Conv2d(img_shape[0]*2, ndf, 4, 2, 1),
            IN(ndf, affine=True), nn.LeakyReLU(0.2, inplace=True),
            nn.Dropout2d(0.3),
        )
        self.conv2 = nn.Sequential(
            nn.Conv2d(ndf, ndf*2, 4, 2, 1),
            IN(ndf*2, affine=True), nn.LeakyReLU(0.2, inplace=True),
            nn.Dropout2d(0.3),
        )
        self.out = nn.Sequential(
            nn.Conv2d(ndf*2, 1, 16), nn.Sigmoid()
        )
        self.img_shape = img_shape

    def forward(self, img, cond_emb, return_features=False):
        batch_size = img.size(0)
        emb_proj = self.embed_fc(cond_emb).view(batch_size, 1, self.img_shape[1], self.img_shape[2])
        emb_proj = emb_proj.expand(-1, img.shape[1], -1, -1)
        x = torch.cat([img, emb_proj], dim=1)
        f1 = self.conv1(x)
        f2 = self.conv2(f1)
        out = self.out(f2)
        out = out.mean([2,3])
        if return_features:
            return out, f2
        else:
            return out

def initialize_weights(m):
    if isinstance(m, (nn.Conv2d, nn.ConvTranspose2d, nn.Linear)):
        nn.init.normal_(m.weight, mean=0.0, std=0.02)
        if m.bias is not None:
            nn.init.constant_(m.bias, 0)

generator = ConditionalDCGANGenerator(latent_dim, embedding_dim, ngf, img_shape).to(device)
discriminator = ConditionalDCGANDiscriminator(embedding_dim, ndf, img_shape).to(device)
generator.apply(initialize_weights)
discriminator.apply(initialize_weights)

criterion = nn.BCELoss()
g_optimizer = torch.optim.Adam(generator.parameters(), lr=1e-4, betas=(0.5, 0.999))
d_optimizer = torch.optim.Adam(discriminator.parameters(), lr=5e-5, betas=(0.5, 0.999))

feature_matching_weight = 25.0  # Set higher if G overpowers D, lower if too weak

epochs = 500
sample_interval = 50
g_losses, d_losses = [], []

def pixel_accuracy(img1, img2, threshold=0.5):
    img1 = ((img1 + 1) / 2).round()
    img2 = ((img2 + 1) / 2).round()
    return (img1 == img2).float().mean().item()

for epoch in range(epochs):
    epoch_g_loss, epoch_d_loss = 0.0, 0.0
    for batch_idx, (batch_imgs, batch_embeddings) in enumerate(loader):
        batch_imgs, batch_embeddings = batch_imgs.to(device), batch_embeddings.to(device)
        batch_size = batch_imgs.size(0)
        real = torch.ones(batch_size, 1).to(device) * 0.95
        fake = torch.zeros(batch_size, 1).to(device)
        noisy_imgs = batch_imgs + torch.randn_like(batch_imgs) * 0.05

        # --- Train Discriminator ---
        d_real, feat_real = discriminator(noisy_imgs, batch_embeddings, return_features=True)
        noise = torch.randn(batch_size, latent_dim).to(device)
        fake_imgs = generator(noise, batch_embeddings)
        d_fake, feat_fake = discriminator(fake_imgs.detach(), batch_embeddings, return_features=True)

        d_loss = criterion(d_real, real) + criterion(d_fake, fake)
        d_optimizer.zero_grad()
        d_loss.backward()
        d_optimizer.step()

        # --- Train Generator (with feature matching) ---
        noise = torch.randn(batch_size, latent_dim).to(device)
        fake_imgs = generator(noise, batch_embeddings)
        g_fake, feat_gfake = discriminator(fake_imgs, batch_embeddings, return_features=True)
        g_loss = criterion(g_fake, real)
        # Feature matching loss (L1 on mean feature maps, across batch)
        match_loss = torch.nn.functional.l1_loss(feat_gfake.mean(0), feat_real.detach().mean(0))
        g_total_loss = g_loss + feature_matching_weight * match_loss
        g_optimizer.zero_grad()
        g_total_loss.backward()
        g_optimizer.step()

        epoch_g_loss += g_total_loss.item()
        epoch_d_loss += d_loss.item()
    g_losses.append(epoch_g_loss)
    d_losses.append(epoch_d_loss)
    print(f"Epoch {epoch+1}: D loss={epoch_d_loss:.3f}, G loss={epoch_g_loss:.3f}")

    if (epoch + 1) % sample_interval == 0 or epoch == 0:
        generator.eval()
        with torch.no_grad():
            shapes = sorted(list(set(all_labels)))
            label_to_idx = {shape: all_labels.index(shape) for shape in shapes}
            sample_idxs = [label_to_idx[shape] for shape in shapes]
            sample_noise = torch.randn(len(sample_idxs), latent_dim).to(device)
            sample_cond_emb = torch.tensor(embeddings[sample_idxs], dtype=torch.float32).to(device)
            samples = generator(sample_noise, sample_cond_emb).cpu().numpy()
            accs = []
            for idx, i in enumerate(sample_idxs):
                real = augment(standardize_fg_bg(Image.open(img_paths[i]).convert('L'))).numpy()
                gen = samples[idx]
                accs.append(pixel_accuracy(torch.tensor(gen), torch.tensor(real)))
            print("Validation pixel accuracy on shapes: %.3f +/- %.3f" % (np.mean(accs), np.std(accs)))
        plt.figure(figsize=(14,2))
        for i, idx in enumerate(sample_idxs):
            img = (samples[i][0] + 1) / 2
            plt.subplot(1, len(sample_idxs), i+1)
            plt.imshow(img, cmap='gray')
            plt.title(all_labels[idx], fontsize=10)
            plt.axis('off')
        plt.suptitle(f"DCGAN-CGAN One Sample per Shape - Epoch {epoch+1}")
        plt.tight_layout()
        plt.show()
        generator.train()

plt.figure(figsize=(8,4))
plt.plot(range(1, len(g_losses)+1), g_losses, label='Generator Loss')
plt.plot(range(1, len(d_losses)+1), d_losses, label='Discriminator Loss')
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("DCGAN Conditional Generator & Discriminator Loss")
plt.legend()
plt.tight_layout()
plt.show()
```
